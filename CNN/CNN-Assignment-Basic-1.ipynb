# What is CNN ?


#### A CNN is a neural network that typically contains several types of layers, which includes :-

* convolutional layer 
* activation layer, and 
* pooling layer.

## Convolutional Layer

* To understand what a CNN is, you need to understand how convolutions work. 

* Imagine you have an image represented as a 5x5 matrix of values, and you take a 3x3 matrix and slide that 3x3 window around the image. 

* At each position the 3x3 visits, you matrix multiply the values of your 3x3 window by the values in the image that are currently being covered by the window. 

* This results in a single number the represents all the values in that window of the image.

```python
def compute_convolution(matrix,convolution,submatrix, stride,bias):
    matrows = matrix.shape[0]
    matcols = matrix.shape[1]
    convrows = convolution.shape[0]
    convcols = convolution.shape[1]
    output_rows = ((matrows-convrows)//stride)+1
    output_cols = ((matcols-convcols)//stride)+1
    temp = []
    for num in submatrix:
        q = ((num*convolution)+bias).sum()
        temp.append(q)
    temp = np.array(temp)
    conv = temp.reshape(output_rows,output_cols)
    return conv 
```

```python
def submatrix_extraction(matrix, convolution, stride):
    matrows = matrix.shape[0]
    matcols = matrix.shape[1]
    convrows = convolution.shape[0]
    convcols = convolution.shape[1]
    sub_matrix = []
    for row in range(0,matrows,stride):
        if (row+convrows)<=matrows:
            for col in range(0,matcols,stride):
                if (col+convcols)<=matcols:
                    q = matrix[row:row+convrows,col:col+convcols]
                    sub_matrix.append(q)
    sub_matrix = np.array(sub_matrix)
    return sub_matrix      
```

```python
def convolution(matrix,convolution,stride,bias):
    "Returns the convolved matrix"
    height = matrix.shape[0]
    rows = matrix.shape[1]
    cols = matrix.shape[2]
    final = []
    for h in range(height):
        sub_matrix = submatrix_extraction(matrix[h],convolution[h],stride)
        conv_layer = compute_convolution(matrix[h],convolution[h],sub_matrix,stride,bias)
        final.append(conv_layer)
    return np.array(final)
```

## Pooling Layers

* Pooling works very much like convoluting, where we take a kernel and move the kernel over the image, the only difference is the function that is applied to the kernel and the image window isnâ€™t linear.

* Max pooling and Average pooling are the most common pooling functions. 

* Max pooling takes the largest value from the window of the image currently covered by the kernel, while average pooling takes the average of all values in the window.

```python
def max_pooling(matrix,convolution,stride):
    height = matrix.shape[0]
    rows = matrix.shape[1]
    cols = matrix.shape[2]
    final = []
    for h in range(height):
        sub_matrix = submatrix_extraction(matrix[h],convolution[h],stride)
        conv_layer = compute_maxpool(matrix[h],convolution[h],sub_matrix,stride)
        final.append(conv_layer)
    return np.array(final)
```

```python
def compute_maxpool(matrix, convolution, sub_matrix, stride):
    matrows = matrix.shape[0]
    matcols = matrix.shape[1]
    convrows = convolution.shape[0]
    convcols = convolution.shape[1]
    output_rows = ((matrows-convrows)//stride)+1
    output_cols = ((matcols-convcols)//stride)+1
    temp = []
    for sub in sub_matrix:
        q =  np.amax((sub*convolution))
        temp.append(q)
    temp = np.array(temp)
    conv = temp.reshape(output_rows,output_cols)
    return conv
```

```python
def avg_pool(matrix, convolution,stride):
    height = matrix.shape[0]
    rows = matrix.shape[1]
    cols = matrix.shape[2]
    final = []
    for h in range(height):
        sub_matrix = submatrix_extraction(matrix[h],convolution[h],stride)
        conv_layer = compute_avgpool(matrix[h],convolution[h],sub_matrix,stride)
        final.append(conv_layer)
    return np.array(final)
```

```python
def compute_avgpool(matrix, convolution, sub_matrix, stride):
    matrows = matrix.shape[0]
    matcols = matrix.shape[1]
    convrows = convolution.shape[0]
    convcols = convolution.shape[1]
    output_rows = ((matrows-convrows)//stride)+1
    output_cols = ((matcols-convcols)//stride)+1
    temp = []
    for sub in sub_matrix:
        q =  np.average((sub*convolution))
        temp.append(q)
    temp = np.array(temp)
    conv = temp.reshape(output_rows,output_cols)
    return conv
```
